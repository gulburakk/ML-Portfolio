****Lets try to find outliers ****

```python
from sklearn.neighbors import LocalOutlierFactor
```

```python
clf = LocalOutlierFactor(n_neighbors = 40, contamination = 0.1)
```

```python
clf.fit_predict(train_extended)
```

```python
df_scores = clf.negative_outlier_factor_
```

```python
esik_deger = np.sort(df_scores)[40]
```

```python
train_extended_out  = train_extended[df_scores > esik_deger]
```

```python
train_extended_out.info()
```

```python
train_extended_out = train_extended_out.drop("Id"  , axis = 1)
```

```python
TARGET = 'quality'
X = train_extended_out.drop(TARGET , axis =1 )
y = train_extended_out[TARGET]
X_train , X_test , y_train , y_test = train_test_split(X , 
                                                       y, 
                                                       random_state = 12 ,
                                                       test_size =0.33)
```

```python
X_train.head()
```

```python
clf = LazyClassifier(verbose=0,
                     ignore_warnings=True,
                     custom_metric=cohen_kappa_score,
                     predictions=False,
                     random_state=12,
                     classifiers='all')

models, predictions = clf.fit(X_train , X_test , y_train , y_test)
```
